```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#  自己创建数据集
X, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:,0], X[:,1],
           marker="o", s=8)
plt.show()
```

看看这个数据集的分布

```python
color = ["red","pink","orange","gray"]
fig, ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(X[y==i,0],X[y==i,1],
               marker="o",
               s=8,
               c=color[i])
plt.show()
```

基于这个分布， 我们来使用 Kmeans聚类，

```python
from sklearn.cluster import KMeans

n_clusters = 3
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)

y_pred_3 = cluster.labels_
y_pred_3
```

## 聚类算法的模型评估指标

当真实标签未知的时候：轮廓系数

```python
from sklearn.metrics import silhouette_score   # 所有样本轮廓系数的平均值
from sklearn.metrics import silhouette_samples  #所有样本的每个 分值
```

```python
X
```

```python
y_pred
```

```python
silhouette_score(X, y_pred)  # 簇为3 的轮廓系数
```

```python
silhouette_score(X, cluster_4.labels_)   # 簇为4  的 轮廓系数
```

```python
silhouette_score(X,cluster_5.labels_)# 簇为5  的 轮廓系数
```

```python
silhouette_score(X,cluster_6.labels_)# 簇为6  的 轮廓系数
```

 当真实标签未知的时候：Calinski-Harabaz Index

```python
from sklearn.metrics import calinski_harabaz_score
```

```python
calinski_harabaz_score(X, y_pred)
```

```python
calinski_harabaz_score(X, y_pred_3)
```

```python
y_pred_4 = cluster_4.fit_predict(X)
calinski_harabaz_score(X, y_pred_4)
```

```python
y_pred_5 = cluster_5.fit_predict(X)
calinski_harabaz_score(X, y_pred_5)
```

```python
y_pred_6 = cluster_6.fit_predict(X)
calinski_harabaz_score(X, y_pred_6)
```
